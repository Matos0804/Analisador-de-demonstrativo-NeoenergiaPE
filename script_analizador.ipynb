```python
#programa para ler demonstrativo de USFV


import tabula
import pandas as pd
from datetime import datetime, timedelta
#import request

#coletando dados:
    #quantidade de unidades do sistema
qnt_und = int(input('quantas unidades fazem parte da lista de compensação(incluindo geradora)?'))
lista_unidades=[]
for i in range(0,qnt_und):
    if i ==0:
        lista_unidades.append(int(input('Qual conta contrato da Unidade Geradora?')))
    else:
        lista_unidades.append(int(input(f'Qual conta contrato da {i}ª Unidade beneficiária ?')))

print(f'a lista das unidades digitadas foram: {lista_unidades}')
print('Em breve mostraremos o resumo do seu relatório!')

#transformando arquivo PDF em tabela
lista_tabelas = tabula.read_pdf("RelatorioResumo.pdf",pages="all")
print(len(lista_tabelas))




#for tabela in lista_tabelas:
#display(tabela)

tabela_total = []

for i in range (3,len(lista_tabelas)-2):
    tabela_total.append(lista_tabelas[i].copy())
    
display(tabela_total)
print('-=-=-='*30)   
  

tabela2 = tabela_total[0]
tabela3 = tabela_total[1]



#Tratando dados das Tabelas:

    #tratando tabela2
#tabela2 = tabela2[1:]
#tabela2.columns = tabela2.iloc[0]#transformando primeira linha em titulo da tabela
#tabela2 = tabela2[1:]
tabela2.columns = ['Mensal','Cliente','Energia Consumida','Energia Injetada','Energia Faturada','Compensado','Créditos']
tabela2 = tabela2.drop(0,axis = 0)
tabela2 = tabela2.drop(1,axis = 0)




    #tratando tabela3
    
#tentativa 1    
linha_tabela = tabela3.columns #criando variável provisória para receber valor do titulo
tabela3.loc[len(tabela3)] = linha_tabela #criando nova linha no limite inferior da tabela
tabela3.columns = ['Mensal','Cliente','Energia Consumida','Energia Injetada','Energia Faturada','Compensado','Créditos']

#tentativa 2


#juntando tabelas   
tabela = pd.concat([tabela2, tabela3],ignore_index = True) #juntando as duas tabelas 

display(tabela)
#dividindo coluna do mês da tabela
tabela[['Mês','Ano']] = tabela['Mensal'].str.split('/', expand = True)


#substituindo valores do mes em BR para EN
tabela=tabela.replace(['fev','abr','mai','ago','set','out','dez'],['feb','apr','may','aug','sep','oct','dec'])

#juntando coluna mês 
tabela['Mensal'] = tabela['Mês'].str.cat(tabela['Ano'], sep='/')
tabela = tabela.drop('Mês',axis = 1)
tabela = tabela.drop('Ano',axis = 1)


#ajustando formato das variáveis
tabela['Mensal'] = tabela['Mensal'].astype(str)
tabela['Cliente'] = tabela['Cliente'].astype(int)
tabela['Energia Consumida'] = tabela['Energia Consumida'].astype(int)
tabela['Energia Injetada'] = tabela['Energia Injetada'].astype(int)
tabela['Energia Faturada'] = tabela['Energia Faturada'].astype(int)
tabela['Compensado'] = tabela['Compensado'].astype(int)
tabela['Créditos'] = tabela['Créditos'].astype(int)
#tabela['Mensal'] = pd.to_datetime(tabela['Mensal'])



#data_ajustada = datetime.strptime(tabela['Mensal'], '%b/%Y'data_ajustada = datetime.strptime(tabela['Mensal'], '%b/%Y'')
#tabela = datetime.strptime(tabela['Mensal'],'%b/%Y')


#ajustando tabela pelo mes
#tabela = tabela.sort_values(by='Mensal')




#tabela = tabela.sort_values('Mensal', key = lambda x: x['Mensal']) #ordenando os valores

display(tabela)





#analisando Dados
    #pesquisar funçoes para analisar dados
    
    
    #pegar linhas que correspondem a uma condição
tabela_resumo=[]
for i in range (0,qnt_und):
    tabela_resumo.append(tabela.loc[tabela['Cliente'] == lista_unidades[i]].copy())
    display(tabela_resumo[i])

#tabela_gerador = tabela.loc[tabela['Cliente'] == lista_unidades[0]]
#tabela_benef = tabela.loc[tabela['Cliente'] == lista_unidades[1]]
#display(tabela_gerador)
#display(tabela_benef)



#exportando tabelas


tabela.to_excel('dados do memorial.xlsx', index = False)
for i in range (0,qnt_und):
    if i == 0:
        tabela_resumo[i].to_excel(f'tabela resumo geradora.xlsx', index = False)
    else:
        tabela_resumo[i].to_excel(f'tabela resumo beneficiaria {i}.xlsx', index = False)


```
